### 1 - Text preprocessing
Nota: é sempre melhor fazer uma cópia de cada dataframe (i.e., os dados originais) para não se perder ou dar *overwrite* sem querer.

#### Fazer cópias das dataframes *data_pt_df*


*   data_pt_df -> variável "df_pt"
*   data_en_df -> variável "df_en"



```python
df_pt = data_pt_df.copy()
df_pt
```

```python
df_en = data_en_df.copy()
df_en
```

#### *Lowercasing*
O ato de colocar todas as palavras em minúsculas (lowercase).

```python
df_pt['lowercase'] = df_pt['frases'].str.lower()
df_pt
```

```python
df_en['lowercase'] = df_en['frases'].str.lower()
df_en
```

#### Remover pontuação e símbolos especiais
Nota: isto vai remover todo o tipo de pontuação e outros símbolos que não sejam caracteres de letras ou de espaço em branco. Isto pode ter efeitos diferentes dependendo do tipo de caracteres que cada língua usa. Por exemplo, o inglês utiliza o símbolo **'** para expressar diferentes significados e relações entre palavras.

```python
df_pt['remove_ponctuation'] = df_pt['lowercase'].str.replace(r'[^\w\s]', '', regex=True)
df_pt
```

```python
df_en['remove_ponctuation'] = df_en['lowercase'].str.replace(r'[^\w\s]', '', regex=True)
df_en
```

## b) With spaCy

```python
import spacy
!python -m spacy download pt_core_news_sm
```

É preciso fazer load dos modelos pré-treinados do spaCy em todas as línguas que estão a ser pré-processadas:

```python
english = spacy.load('en_core_web_sm')
portuguese = spacy.load('pt_core_news_sm')
```

#### Tokenização

```python
df_pt['tokens'] = df_pt['frases'].apply(portuguese)
df_pt
```

```python
df_en['tokens'] = df_en['frases'].apply(english)
df_en
```

#### Stop Words
Stop Words são palavras  que não adicionam significado semântico relevante para o entendimento do sistema sobre determinado texto.
É preciso dar load também da lista das stops words, uma para cada modelo de língua, uma variável para cada.

```python
stopwords_pt = portuguese.Defaults.stop_words
stopwords_en = english.Defaults.stop_words
```

O spaCy só consegue tirar as stop words com uma string. Assim, é preciso alterar a dataframe para strings. Por agora, usa-se apenas uma das frases para cada língua:

```python
str_pt = df_pt['frases'].iloc[0]
str_pt
```

```python
str_en = df_en['frases'].iloc[0]
str_en
```

Depois de estar no tipo de dados correto, colocar cada frase numa variável (para cada língua):

```python
doc_pt = portuguese(str_pt)
doc_pt
```

```python
doc_en = english(str_en)
doc_en
```

Agora é só fazer a remoção de stop words:

```python
[token.text for token in doc_pt if not token.is_stop]
```

```python
[token.text for token in doc_en if not token.is_stop]
```

#### Lematização
Pode ser feito de forma similar à remoção de stop words:

```python
[token.lemma_ for token in doc_pt]
```

```python
[token.lemma_ for token in doc_en]
```

#### Parts-Of-Speech (POS) tagging
Para isto, é preciso utilizar outro conjunto de frases, visto que o objetivo é dar uma etiqueta (tag) a todas as palavras da frase e não apenas às que são relevantes para o sistema.

Atenção: Cuidado com a sobreposição de variáveis!

```python
doc_pt2 = portuguese(str_pt)
doc_pt2
```

```python
doc_en2 = english(str_en)
doc_en2
```

Agora procedemos ao POS tagging:

```python
[(token.text, token.pos_ ) for token in doc_pt2]
```

```python
[(token.text, token.pos_ ) for token in doc_en2]
```
